```python


# List of categorical columns
categorical_cols = [
    "BRId", "GenderId", "IAId", "Amount of Credit Cards", 
    "Nationality", "Occupation", "Fee Structure", 
    "Loyalty Classification", "Properties Owned", 
    "Risk Weighting", "Income Band"
]

# Loop through each categorical column
for i, col in enumerate(categorical_cols):
    plt.figure(figsize=(10, 6))
    
    if col == "Occupation":  
        # Show only top 10 occupations
        top_10_occupations = df[col].value_counts().head(10).index
        sns.countplot(
            data=df[df[col].isin(top_10_occupations)],
            y=col,
            order=top_10_occupations,
            palette="Set2"
        )
        plt.title(f"Top 10 {col}", fontsize=14)
        plt.xlabel("Count")
        plt.ylabel(col)
    
    else:
        # Sort by frequency for better visualization
        order = df[col].value_counts().index
        sns.countplot(data=df, x=col, order=order, palette="viridis")
        plt.title(f"Distribution of {col}", fontsize=14)
        plt.xlabel(col)
        plt.ylabel("Count")
        plt.xticks(rotation=45)  # Rotate labels if needed
    
    plt.tight_layout()
    plt.show() # as there is no hue, so it is purely a univariate analysis 
```

```python

##Bivariate analysis
```

```python
# List of categorical columns
categorical_cols = [
    "BRId", "GenderId", "IAId", "Amount of Credit Cards", 
    "Nationality", "Occupation", "Fee Structure", 
    "Loyalty Classification", "Properties Owned", 
    "Risk Weighting", "Income Band"
]

# Loop through each categorical column
for i, predictor in enumerate(categorical_cols):
    plt.figure(i, figsize=(10, 6))
    
    if predictor == "Occupation":
        # Limit to top 10 occupations
        top_10_occupations = df["Occupation"].value_counts().head(10).index
        filtered_df = df[df["Occupation"].isin(top_10_occupations)]
        sns.countplot(
            data=filtered_df, 
            x="Occupation", 
            hue="Nationality", 
            order=top_10_occupations,
            palette="Set2"
        )
        plt.xticks(rotation=45)
        plt.title(f"Top 10 Occupations by Nationality")
    else:
        order = df[predictor].value_counts().index
        sns.countplot(
            data=df, 
            x=predictor, 
            hue="Nationality", 
            order=order,
            palette="viridis"
        )
        plt.xticks(rotation=45)
        plt.title(f"Distribution of {predictor} by Nationality")
    
    plt.tight_layout()
    plt.show() #Bivariate analysis happens when we use hue, and split one single column 
    #a certain pattern is observed wrt nationality
```

```python
# HIstplot of value counts for different Occupation

for col in categorical_cols:
  if col == "Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('Histogram of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
```

```python
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts',
    'Saving Accounts', 'Foreign Currency Account', 'Business Lending'
]

# Univariate analysis for numerical columns
plt.figure(figsize=(18, 12))
sns.set_style("whitegrid")

for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i+1)  # 3x3 grid
    sns.histplot(df[col], kde=True, color='skyblue', edgecolor='black')
    plt.title(col, fontsize=12, fontweight='bold')
    plt.xlabel("")
    plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
#Heatmaps to see the correlation between all the numerical attributes
numerical_cols = ['Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python

##Insights of EDA:
The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account type often hold substantial amount/funds across other accounts as well.
a) Financial Health:

A high correlation suggests that individuals with substantial bank deposits are likely to have both checking and savings accounts. This might indicate better overall financial health and stability.

b) Cash Management:

People with more bank deposits may be more proactive in managing their cash flow, utilizing checking accounts for daily transactions while keeping savings accounts for future needs or emergencies.

c) Investment Behavior:

Individuals might be saving more (reflected in higher deposits) and thus might be using checking accounts for regular spending, reflecting a preference for maintaining liquidity.
d) Banking Relationships:

A strong correlation could indicate that customers who use one product (like savings accounts) are likely to use others (like checking accounts) from the same bank, suggesting customer loyalty or bundled banking services.

e) Demographics:

High correlation might also reflect demographic factors where certain groups (age, income level) tend to have similar banking behaviors.

f) Potential for Cross-Selling:

If you're a financial institution, this correlation could indicate opportunities for cross-selling products. Customers with deposits might be more inclined to open additional accounts or services.
```
